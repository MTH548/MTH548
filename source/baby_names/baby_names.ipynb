## Database of baby names

The United States Social Security Administration maintains records of babies born in the USA each year. Historical data of first names of babies is available on the [Social Security website](https://www.ssa.gov/oact/babynames/limits.html). It provides the number of babies registered with a given first name in a given year. Here is a sample data for the year 1960:

```python
# HIDDEN
#  "nbsphinx": "hidden",

import plotly.graph_objects as go
import pandas as pd

year = 1960
df = pd.read_csv(f"names/yob{year}.txt", names=["name", "gender", "count"] ) 

fig = go.Figure(data=[go.Table(
    name=f"Baby names in {year}",
    columnwidth = 8,
    header=dict(values=df.columns,
                font_color='white',
                fill_color='royalblue',
                align='left'),
    cells=dict(values= [df[c] for c in df.columns],
               fill_color='lavender',
               align='left'))
])

fig.update_layout(title_text=f"USA baby names in {year}", 
                  title_font_color="Black",
                  title_font_size=18,
                  width=600,
                  height=400)

fig.show()
fig.write_html(f"../_static/baby_names/year_names.html", include_plotlyjs='cdn')
```

Another data set provides the number of babies registered with a given first name in each US state. Here is a sample of data for New York:

```python
# HIDDEN
#  "nbsphinx": "hidden",

state = "NY"
df = pd.read_csv(f"namesbystate/{state}.txt", names=["state", "gender", "year",  "name", "count"])[:10000] 

fig = go.Figure(data=[go.Table(
    name=f"Baby names in {year}",
    columnwidth = 8,
    header=dict(values=df.columns,
                font_color='white',
                fill_color='royalblue',
                align='left'),
    cells=dict(values= [df[c] for c in df.columns],
               fill_color='lavender',
               align='left'))
])

fig.update_layout(title_text=f"Names of babies born in {state}", 
                  title_font_color="Black",
                  title_font_size=18,
                  width=600,
                  height=400)

fig.show()
fig.write_html("../_static/baby_names/state_names.html", include_plotlyjs='cdn')
```
