## k-NN implementations

A function implementing k-NN fitting. Distances between data vectors are measured using $l^p$ metric, where $p$ can be specified. 

```python
def knn(training, labels, data, k, p=2):
    """
    training: 
        A 2-dimensional array with training data (record in rows).
    labels:
        A 1-dimensional array with training data labels.
    data:
        An array with a data point to classify.
    k:
        The number of neighbors to use.
    p:
        Exponent for the Lp metric
    
    Returns:
        
    unique: 
        An array with labels of neighbors, ordered from the label 
        with the biggest number of occurencies to the label with 
        the smallest number. 
    
    counts:
       An array with number of occurencies of each unique label, 
       in the same order as unique. 
            
    neighbor_rows:
       Indices of rows of neighbors in the training array.    
    """
    
    
    # compute distances between training set and the data point
    distances = (np.abs(training - data.ravel())**p).sum(axis=1)
    
    # get row numbers of the nearest neighbors
    neighbor_rows = np.argpartition(distances, k-1)[:k]
    # get labels of the neighbors
    neighbor_labels = labels[neighbor_rows]
    # get unique labels and their counts among neighbors
    uniques, counts = np.unique(neighbor_labels, return_counts=True)
    # indexes of counts in descending order
    ranks = np.argsort(counts)[::-1]
    return uniques[ranks], counts[ranks], neighbor_rows
```

Test accuracy of classification based on various numbers of neighbors:

```python
%%timeit -n1 -r1

tot = 500

for k in range(1, 8):
    s = 0
    for n in range(tot):
        l = test_labels[n]
        p, *_ = knn(training_pics, training_labels, data = test_pics[n], k=k, p=2)
        if l==p[0]: s+=1
    print(f"k={k}, accuracy={s/tot}")
```

Here is a somewhat simpler implementation of k-NN:

```python
def knn_easy(training, labels, data, k, p=2):
    
    distances = (np.abs(training - data.ravel())**p).sum(axis=1)
    
    # get row numbers of the nearest neighbors
    neighbors = np.argsort(distances)[:k]
    # get labels of the neighbors
    neighbor_labels = labels[neighbors]
    
    counts = {}
    for n in neighbor_labels:
        if n in counts:
            counts[n] += 1
        else:
            counts[n] = 1
            
    return max(counts.items(), key = lambda x: x[1])[0], counts 
```

Testing accuracy of the above implementation. The results in both cases differ a bit, since in each case we do not specify explicitly how to select label in case of a tie between neighbors in various classes. 

```python
%%timeit -n1 -r1

tot = 500

for k in range(1, 8):
    s = 0
    for n in range(tot):
        l = test_labels[n]
        p = knn_easy(training_pics, training_labels, data = test_pics[n], k=k, p=2)
        if l==p[0]: s+=1
    print(f"k={k}, accuracy={s/tot}")
```

Testing accuracy with the k-NN implementation provided by sklearns. The sklearn implementation is much faster, since it uses a better method of searching for neighbors.

```python
%%timeit -n1 -r1

from sklearn.neighbors import KNeighborsClassifier

tot = 500

for k in range(1, 8):
    neigh = KNeighborsClassifier(n_neighbors=k, p=2)
    neigh.fit(training_pics, training_labels)
    predictions = neigh.predict(test_pics[:tot])
    print(f"k={k}, accuracy={(np.array(predictions) == test_labels[:tot]).sum()/tot}")
```

## Analizing results

Lets have a look at the digits that k-NN misclassifies. First, we compute predictions for 500 number of test samples using 5 neighbors:

```python
tot = 500
k = 5
results = []
for n in range(tot):
    l = test_labels[n]
    unique, counts, neighbors = knn(training_pics, training_labels, data = test_pics[n], k=k, p=2)
    results.append((n, l, unique, counts, neighbors))
```

Next we plot each misclassified digit with its actual and predicted labels. Below it, we plot its neighbors used to make the prediction. The next row of plots shows the difference between the misclassified digit and each of its neighbor, to better illustrate the similarity between these images: 

```python
for n, l, unique, counts, neighbors in results:
    if l!=unique[0]:
        print(f"label:{l}, prediction:{unique[0]}")
        show_digit(test_pics[n])
        
        w, h = 28,28
        fig = plt.figure(figsize=(3.5,2))
        plt.style.use('seaborn-white')
        for i in range(k):
            plt.subplot(2, k, i+1)
            plt.xticks([])
            plt.yticks([])
            plt.title(training_labels[neighbors[i]])
            plt.imshow(training_pics[neighbors[i]].reshape(h,w), cmap="gray_r")
            plt.subplot(2, k, k+i+1)
            plt.xticks([])
            plt.yticks([])
            plt.imshow((training_pics[neighbors[i]].reshape(h,w) - test_pics[n].reshape(h,w))**2, cmap="viridis_r")
        plt.show()
```

## Confusion matrix

Compute and display a confusion matrix for based on the 500 predictions computed above:

```python
lab = [r[1] for r in results]
pred = [r[2][0] for r in results]
conf = np.zeros((10,10))
np.add.at(conf, (lab, pred), 1)

conf
```
