```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

## Table scrapping with `pandas`

```python
url = "http://www.buffalo.edu/class-schedule?switch=showcourses&semester=spring&division=1&dept=MTH"
```

```python
tables = pd.read_html(url)
tables[4]
```

## Table scrapping with Beautiful Soup

```python
page = requests.get(url).text
```

```python
soup = BeautifulSoup(page)
tables = soup.find_all('table')
course_table = tables[5]
rows =  course_table.find_all("tr")

data = []
for row in rows[4:]:
    d = [s.get_text().strip() for s in row.find_all("td")]
    if "Class" not in d[0]:
        link = "" if row.find("a") is None else row.find("a")['href']
        d.append(link)
        data.append(d)
headers = list(rows[3].stripped_strings)
headers.append("Link")

df = pd.DataFrame(data, columns = headers)
df
```

```python
df['Link'][10]
```

Add interactive links:

```python
def make_clickable(val):
    # target _blank to open new window
    return '<a target="_blank" href="{}">{}</a>'.format(val, val.replace("&", "&amp;"))

df.iloc[:10].style.format({'Link': make_clickable})
```

## Searching for enrollment information

```python
def get_enrollment(url):
    s = BeautifulSoup(requests.get(url).text)
    cap =  [td for td in s.find_all('td') if "Enrollment Capacity:" in td.get_text()][0].find_next('td')
    tot =  [td for td in s.find_all('td') if "Enrollment Total:" in td.get_text()][0].find_next('td')
    cap = int(cap.get_text())
    tot = int(tot.get_text())
    return cap, tot
```

```python
for i in range(5):
    display(df.iloc[i:i+1])
    print(get_enrollment(df['Link'][i]))
```
