## Counting words

Counting words with a dictionary:

```python
count_words = {}
for w in words:
    if w not in count_words:
        count_words[w] = 1
    else:
        count_words[w] += 1
count_words
```

```python
count_words = {}
for w in words:
    count_words[w] = count_words.get(w, 0) + 1
count_words
```

Counting words with a counter:

```python
from collections import Counter

c1 = Counter(words)
c1
```

```python
c1["this"]
```

```python
c1["buffalo"]
```

```python
c1.most_common()
```

```python
sum(c1.values())
```

```python
len(c1)
```

## Split reviews into training and testing data

```python
movies.head(5)
```

```python
train_frac = 0.5
train_df = movies.sample(frac=train_frac, random_state=0).copy()
train_df
```

```python
test_df = movies.drop(train_df.index).copy()
test_df
```

```python
def wc(text):
    words = re.findall("[a-z']+", text.lower().replace("<br />", ""))
    return Counter(words)
```

```python
train_df.loc[14149, "review"]
```

```python
wc(train_df.loc[14149, "review"])
```

```python
train_df["word_counts"] = train_df["review"].map(wc)
train_df
```

This works fine but it is slow:

```python
train_df.groupby(by="sentiment")["word_counts"].sum()
```

A faster way:

```python
train_df["sentiment"].unique()
```

```python
counters = {}
for label in train_df["sentiment"].unique():
    counters[label] = Counter()

for i in train_df.index:
    counters[train_df.loc[i, "sentiment"]] += train_df.loc[i, "word_counts"]
```

```python
word_counts = pd.DataFrame(counters)
word_counts
```

```python
word_counts = word_counts.fillna(0)
word_counts
```

```python
word_counts.sort_values(by="negative", ascending=False)
```

```python
word_counts.sort_values(by="positive", ascending=False)
```

```python
with open("stopwords.txt") as foo:
    stops = foo.read().split(",")
print(stops)
```

```python
word_counts = word_counts.drop(stops, errors="ignore")
```

```python
word_counts.sort_values(by="negative", ascending=False)
```
